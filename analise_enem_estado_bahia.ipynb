**diferenças de notas entre escola publica e privada**


```python
df.query('TP_ESCOLA == 2')['NU_NOTA_MT'].mean()
```

```python
df.query('TP_ESCOLA == 3')['NU_NOTA_MT'].mean()
```

é possivel concluir que escola privada tem media superior a media das escola publica mesmo tendo poucas escolas privada na região

```python
df.query('TP_ESCOLA == 2').NU_NOTA_MT.agg([np.mean, np.median])
```
